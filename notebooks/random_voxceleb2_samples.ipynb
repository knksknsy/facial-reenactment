```python
data_frame = data_frame[~data_frame['id'].isin(ids)]
data_frame.shape
```

```python
unique_samples_id = data_frame.drop_duplicates('id')
unique_samples_id.shape
```

```python
unique_ids = unique_samples_id['id'].tolist()
unique_df = pd.DataFrame(columns=['landmark1','image1','landmark2','image2','landmark3','image3','id','id_video'])

for idx in unique_ids:
    sample = data_frame[data_frame['id'] == idx].sample(2)
    sample = sample.drop(sample.columns[list(range(6,18))], axis=1)
    unique_df = unique_df.append(sample, ignore_index=True)
```

```python
unique_df = unique_df.sort_values(by=['id'], ascending=True)
unique_df.shape
```

```python
unique_df_even = unique_df.iloc[::2]
unique_df_odd = unique_df.iloc[1::2]
```

```python
unique_df_odd = unique_df_odd.sample(frac=1)
unique_df_odd = unique_df_odd[0:3823]
unique_df_odd = unique_df_odd.sort_values(by=['id'], ascending=True)
```

```python
unique_df = pd.concat([unique_df_even, unique_df_odd])
unique_df = unique_df.sort_values(by=['id'], ascending=True)
unique_df.shape
```

```python
unique_df.to_csv('dataset_unique.csv', sep=',', index=False)
```
